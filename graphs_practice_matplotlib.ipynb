```python
#line graph
import matplotlib.pyplot as plt
import numpy as np

xpoints = np.array([0, 6,8,6,4,9])
ypoints = np.array([0, 250,60,80,50,10])

plt.plot(xpoints, ypoints)
plt.show()
```

```python
#bar graph
import matplotlib.pyplot as plt
import numpy as np

x = np.array(["CSE", "MATH1", "MATH2", "PHYSICS"])
y = np.array([98, 88, 78, 72])

plt.bar(x,y)
plt.show()
```

```python
#pie
import matplotlib.pyplot as plt
import numpy as np

y = np.array([98, 88, 78, 72])
a=["csen","math1","math2","physics"]
plt.pie(y,labels=a)
plt.show()
```

```python
#scatter(points)
import matplotlib.pyplot as plt
import numpy as np
x = np.array(["CSE", "MATH1", "MATH2", "PHYSICS"])
y = np.array([98, 88, 78, 72])



plt.scatter(x, y)
plt.show()
```

```python
#grid graph
import numpy as np
import matplotlib.pyplot as plt

x = np.array(["CSE", "MATH1", "MATH2", "PHYSICS"])
y = np.array([98, 88, 78, 72])

plt.title("report")
plt.xlabel("subject")
plt.ylabel("marks")

plt.plot(x,y)

plt.grid()


plt.show()
```

```python
##histogram
import matplotlib.pyplot as plt
import numpy as np

x = np.random.normal(170, 10, 180)

plt.hist(x)
plt.show() 

```

```python
#pie
import matplotlib.pyplot as plt
import numpy as np

y = np.array([98, 88, 78, 72])
a=["csen","math1","math2","physics"]
e = [0,0.2,0,0]
plt.title("LANSUM",color = "r")
plt.pie(y,labels=a,explode = e)
plt.show()
```
